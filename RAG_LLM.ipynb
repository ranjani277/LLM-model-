```python
text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=100)
texts = text_splitter.split_documents(documents)
texts
```

```python
type(texts)
```

```python
texts[2]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = 'vector_db'
vectordb = Chroma.from_documents(texts, embeddings, persist_directory=persist_directory)
```

```python
retrieval = vectordb.as_retriever()
```

```python

llm = ChatGroq(model = "Llama-3.1-70b-Versatile",temperature=0)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retrieval,return_source_documents=True)
```

```python
query = "what are the inbuild function in python?"
result = qa_chain({"query": query})
result["result"]
```
